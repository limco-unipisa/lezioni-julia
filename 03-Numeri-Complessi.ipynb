# 3. Traslazioni e rotazioni sul piano

Queste trasformazioni si effettuano con l'aritmetica complessa. Dato il vettore rappresentato dal numero complezzo $z$:

*   Traslazione di $w$: $z\rightarrow z + w$
*   Rotazione di angolo $\theta$: $z\rightarrow z \cdot e^{i\theta}$

**Esempio.** Per ruotare di un angolo $\pi/4$ il quadrato con vertici definiti dal vettore
`quadrato = [0+0im, 1+0im, 1+1im, 0+1im]` e traslarlo di un vettore `v = 3+2i` attorno al baricentro si possono dare i seguenti comandi

```julia
quadrato = [0+0im, 1+0im, 1+1im, 0+1im]
v = 3 +2im
t = pi/4
ruotato = quadrato.* exp(im * t) # Applica rotazione di π/4
traslato_ruotato = ruotato .+ v  # Applica traslazione

# aggiunge il primo elemento per chiudere il poligono
quadrato = push!(quadrato, quadrato[1])
traslato_ruotato = push!(traslato_ruotato, traslato_ruotato[1])

# Disegna il quadrato originale e quello trasformato
plot(real.(quadrato), imag.(quadrato), label="Originale", aspect_ratio=:equal)
plot!(real.(traslato_ruotato), imag.(traslato_ruotato), label="Trasformato", color=:red)

````julia
quadrato = [0+0im, 1+0im, 1+1im, 0+1im]
v = 3 +2im
t = pi/4
ruotato = quadrato.* exp(im * t) # Applica rotazione di π/4
traslato_ruotato = ruotato .+ v  # Applica traslazione

# aggiunge il primo elemento per chiudere il poligono
quadrato = push!(quadrato, quadrato[1])
traslato_ruotato = push!(traslato_ruotato, traslato_ruotato[1])

# Disegna il quadrato originale e quello trasformato
plot(real.(quadrato), imag.(quadrato), label="Originale", aspect_ratio=:equal)
plot!(real.(traslato_ruotato), imag.(traslato_ruotato), label="Trasformato", color=:red)
````
